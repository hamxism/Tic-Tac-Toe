```python
def tic_tac(position):
    print("   |   |")
    print(" " + position[0] + " | " + position[1] + " | " + position[2])
    print("   |   |")
    print("____________")
    print("   |   |")
    print(" " + position[3] + " | " + position[4] + " | " + position[5])
    print("   |   |")
    print("____________")
    print("   |   |")
    print(" " + position[6] + " | " + position[7] + " | " + position[8])
    print("   |   |")

def start_game():
    position = [" ", " ", " ", " ", " ", " ", " ", " ", " "]
    user = "X"
    game_over = False
    tic_tac(position)
    while not game_over:
        if user == "X": # User turn denoting  X's 
            move = int(input("Input number from 1 to 9: "))
            if position[move-1] == " ":
                position[move-1] = user
                if winning(position, user):
                    print("User denoting ",user, "wins!")
                    game_over = True
                elif tie(position):
                    print("Game Tie!!!!!!!!!!!!!")
                    game_over = True
                else:
                    user = "O"
            else:
                print("Position Already Aquired!!!!!!")
        else:# Computer's turn denoting 'O'
            move = computer_turn(position, user)
            position[move] = user
            print("The computer chooses position", move+1)
            tic_tac(position)
            if winning(position, user):
                print("Congrats!!!!!!")
                print("User denoting ",user, "wins!")
                game_over = True
            elif tie(position):
                print("It's a tie!")
                game_over = True
            else:
                user = "X"

def computer_turn(position, user):
    available_moves = [i for i, x in enumerate(position) if x == " "]
    best_move = None
    best_score = -1000
    for move in available_moves:
        position[move] = user
        score = minmax(position, "X" if user == "O" else "O")
        position[move] = " "
        if score > best_score:
            best_score = score
            best_move = move
    return best_move

def minmax(position, user):
    if winning(position, "O"):
        return -1                     #utility value=-1
    elif winning(position, "X"):
        return 1                      #utility value=1
    elif tie(position):
        return 0                      #utility value=0
    available_moves = [i for i, x in enumerate(position) if x == " "]

    if user == "X":# 'X' wants to minimize the score
        best_score = float("inf")
        for move in available_moves:
            position[move] = user
            score = minmax(position, "O")
            position[move] = " "
            if score < best_score:
                best_score = score
        return best_score
    else: # 'O' wants to maximize the score
        best_score = -float("inf")
        for move in available_moves:
            position[move] = user
            score = minmax(position, "X")
            position[move] = " "
            if score > best_score:
                best_score = score
        return best_score
def winning(position, user):
    win_positions = [[0, 1, 2], [3, 4, 5], [6, 7, 8],[0, 3, 6], [1, 4, 7], [2, 5, 8],[0, 4, 8], [2, 4, 6] ]
    for positions in win_positions:
        if all(position[p] == user for p in positions):
            return True
    return False

def tie(position):

    return all(x != " " for x in position)


print("Starting Game")
start_game()
```
